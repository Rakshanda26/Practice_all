```python
import pandas as pd

# Sample DataFrame in wide format
data = {'Date': ['2022-01-01', '2022-01-02', '2022-01-03'],
        'Apple_Price': [2.5, 2.7, 2.9],
        'Banana_Price': [1.0, 1.2, 1.5],
        'Orange_Price': [3.0, 2.8, 3.2]}

wide_df = pd.DataFrame(data)

# Reshaping from wide to long format using melt
long_df = pd.melt(wide_df, id_vars=['Date'], var_name='Fruit', value_name='Price')

# Displaying the long format DataFrame
print("DataFrame in Long Format:")
print(long_df)

# Reshaping from long to wide format using pivot
wide_df_pivot = long_df.pivot(index='Date', columns='Fruit', values='Price').reset_index()

# Displaying the wide format DataFrame after pivot
print("\nDataFrame in Wide Format (after pivot):")
print(wide_df_pivot)
```

```python
import pandas as pd
data = {'Date': ['2022-01-01', '2022-02-15', '2022-03-20', '2022-04-10'],
        'Value': [25, 30, 22, 35]}

df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Is_Weekend'] = df['Date'].dt.weekday//5
df['Quarter'] = df['Date'].dt.quarter
df['Is_Leap_Year'] = df['Date'].dt.is_leap_year
df['Day_since_today'] = (pd.to_datetime('today') - df['Date']).dt.days
df['Start_of_Month'] = df['Date'] - pd.offsets.MonthBegin(0)
df['Start_of_Month'] = df['Date'] + pd.offsets.MonthBegin(0)


```

```python
import numpy as np 
#Fills NaN values in one DataFrame with non-NaN values from another DataFrame.
# Sample DataFrames for combining
df1 = pd.DataFrame({'A': [1, 2, np.nan],
                    'B': [4, np.nan, 6],
                    'C': [7, 8, 9]})

df2 = pd.DataFrame({'A': [np.nan, 2, 3],
                    'B': [4, 5, np.nan],
                    'C': [10, 11, 12]})

# Combine values from df2 where df1 has NaN values
result_combine = df1.combine_first(df2)
print("\nCombining DataFrames:")
print(result_combine)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame for combining column-wise
df = pd.DataFrame({'A': [1, 2, np.nan],
                   'B': [4, np.nan, 6],
                   'C': [7, 8, 9],
                   'D': [10, 11, 12]})

# Combine values from 'D' where 'A' has NaN values
df['A'] = df['A'].fillna(df['D'])

# Displaying the result
print("Combined DataFrame:")
print(df)
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame for combining column-wise
df = pd.DataFrame({'A': [1, 2, np.nan],
                   'B': [4, np.nan, 6],
                   'C': [7, 8, 9],
                   'D': [10, 11, np.nan]})

# Combine values from 'D' where 'A' has NaN values
df['A'] = df['A'].combine_first(df['D'])

# Displaying the result
print("Combined DataFrame:")
print(df)
```

```python
# Sample DataFrame for appending column-wise
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2']})

df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2'],
                    'D': ['D0', 'D1', 'D2']})

# Append one DataFrame to another column-wise
result_append = pd.concat([df1, df2], axis=1)
print("\nAppending DataFrames column-wise:")
print(result_append)
```

```python
# Sample DataFrame for merging column-wise
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                    'value1': ['V0', 'V1', 'V2']})

df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                    'value2': ['V3', 'V4', 'V5']})

# Merge based on the 'key' column
result_merge = pd.merge(df1, df2, on='key', how='inner')
print("\nMerging DataFrames column-wise:")
print(result_merge)
```
